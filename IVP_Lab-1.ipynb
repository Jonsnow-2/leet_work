```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('image_new.jpeg')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
_, albedo_mask = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)
result = cv2.bitwise_and(img, img, mask=albedo_mask)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
```

```python
plt.imshow(albedo_mask, cmap='gray'), plt.title('Albedo Mask')
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB)), plt.title('Result')
```

```python
albedo_mask_3ch = cv2.cvtColor(albedo_mask, cv2.COLOR_GRAY2BGR)
result = cv2.bitwise_and(img, 255 - albedo_mask_3ch)
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB)), plt.title('Result')
```
