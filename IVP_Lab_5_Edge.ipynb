```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('Butterfly.jpg', cv2.IMREAD_GRAYSCALE)
```

## (i) Roberts

```python
roberts_x = np.array([[1, 0],
                      [0, -1]])
roberts_y = np.array([[0, 1],
                      [-1, 0]])

roberts_edges_x = cv2.filter2D(image, -1, roberts_x)
roberts_edges_y = cv2.filter2D(image, -1, roberts_y)

roberts_edges = np.sqrt(np.square(roberts_edges_x) + np.square(roberts_edges_y))
```

## (ii) Sobel

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

sobel_edges = np.sqrt(np.square(sobel_x) + np.square(sobel_y))
```

## (iii) Prewitt

```python
prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])
prewitt_y = np.array([[-1, -1, -1],
                      [0, 0, 0],
                      [1, 1, 1]])

prewitt_edges_x = cv2.filter2D(image, -1, prewitt_x)
prewitt_edges_y = cv2.filter2D(image, -1, prewitt_y)

prewitt_edges = np.sqrt(np.square(prewitt_edges_x) + np.square(prewitt_edges_y))
```

## (iv) Laplacian

```python
laplacian_edges = cv2.Laplacian(image, cv2.CV_64F)
laplacian_edges = np.uint8(np.absolute(laplacian_edges))
```

## (v) Canny

```python
canny_edges = cv2.Canny(image, 100, 200)
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(roberts_edges, cmap='gray'), plt.title('Roberts Operator')
plt.subplot(2, 3, 3), plt.imshow(sobel_edges, cmap='gray'), plt.title('Sobel Operator')
plt.subplot(2, 3, 4), plt.imshow(prewitt_edges, cmap='gray'), plt.title('Prewitt Operator')
plt.subplot(2, 3, 5), plt.imshow(laplacian_edges, cmap='gray'), plt.title('Laplacian Operator')
plt.subplot(2, 3, 6), plt.imshow(canny_edges, cmap='gray'), plt.title('Canny Operator')

plt.show()
```
