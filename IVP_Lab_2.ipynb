```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('image.jpg')
```

```python
img.shape
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

# (i) Resizing Images

```python
width, height = 300,150
resized_nearest = cv2.resize(img, (width, height), interpolation=cv2.INTER_NEAREST)
```

```python
rgb_img = cv2.cvtColor(resized_nearest, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
resized_bilinear = cv2.resize(img, (width, height), interpolation=cv2.INTER_LINEAR)
rgb_img = cv2.cvtColor(resized_bilinear, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
resized_bicubic = cv2.resize(img, (width, height), interpolation=cv2.INTER_CUBIC)
rgb_img = cv2.cvtColor(resized_bicubic, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

# (ii) Conversion to different formats

```python
cv2.imwrite('output_image.png', img)
```

```python
cv2.imwrite('output_image.bmp', img)
```

```python
cv2.imwrite('output_image.jpeg', img, [cv2.IMWRITE_JPEG_QUALITY, 90])
```

# (iii) Extraction of Color bands from Image

```python
blue_band = img[:, :, 0]
green_band = img[:, :, 1]
red_band = img[:, :, 2]
```

```python
cv2.imwrite('blue_band.jpg', blue_band)
cv2.imwrite('green_band.jpg', green_band)
cv2.imwrite('red_band.jpg', red_band)
```

```python
blue_band = img[:, :, 0]
green_band = img[:, :, 1]
red_band = img[:, :, 2]

# Visualize each color band
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(blue_band, cmap='Blues')
plt.title('Blue Band')

plt.subplot(132)
plt.imshow(green_band, cmap='Greens')
plt.title('Green Band')

plt.subplot(133)
plt.imshow(red_band, cmap='Reds')
plt.title('Red Band')

plt.tight_layout()
plt.show()
```
